# Python Tuples
The main differences between lists and tuples are: Lists are enclosed in brackets ( [ ] ) and their elements and size can be changed, while tuples are enclosed in parentheses ( ( ) ) and cannot be updated. Tuples can be thought of as read-only lists

```python
tuple = ( 'abcd', 786 , 2.23, 'john', 70.2  )
tinytuple = (123, 'john')

print (tuple)           # Prints all items in Tuple
print (tuple[0])        # Prints first element of the Tuple
print (tuple[1:3])      # Prints elements starting from 2nd till 3rd 
print (tuple[2:])       # Prints elements starting from 3rd element
print (tinytuple * 2)   # Prints list two times
print (tuple + tinytuple) # Prints concatenated lists
```

# File Handling in Python (Most Crucial for Data Science)
The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

# File Types also needs to be mentioned

"t" - Text - Default value. Text mode

"b" - Binary - Binary mode (e.g. images)

```python
f = open("myFile.txt", "rt")
print(f)
```

```python
f = open("myFile.txt", "r")
print(f.read())
```

```python
# Return the 5 first characters of the file:

f = open("myFile.txt", "r")
print(f.read(25))
```

```python
# Open file and append file contents accordingly

f = open("demofile2.txt", "a")
f.write("NEW CONTENT HAS BEEN GENERATED!")
f.close()

#open and read the file after the appending:
f = open("demofile2.txt", "r")
print(f.read())
f.close()
```

# Object Oriented in Python

```python
class Addition:
    # Declaring Data Members of class Addition
    first = 0
    second = 0
    answer = 0
    # parameterized constructor 
    def __init__(self, f, s): 
        self.first = f 
        self.second = s 
        
    def display(self): 
        print("First number = " + str(self.first)) 
        print("Second number = " + str(self.second)) 
        print("Addition of two numbers = " + str(self.answer)) 
  
    def calculate(self): 
        self.answer = self.first + self.second 

# creating object of the class 
# this will invoke parameterized constructor 
obj = Addition(1000, 2000) 
  
# perform Addition 
obj.calculate() 
  
# display result 
obj.display()
```
